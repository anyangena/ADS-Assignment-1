## Question 4 : Cleaning the data set - part B 




## a) Run the 'alumni["Gender"].value_counts()' to see the incorrect 'M' fields that need to be converted to 'Male'

```python
alumni['Gender'].value_counts()
```

## Now use a '.str.replace' on the 'Gender' column to covert the incorrect 'M' fields. Hint: We must use ^...$ to restrict the pattern to match the whole string.

```python
alumni['Gender'].str.replace("^M$","Male")
```

```python
alumni
```

## c) That didn't the set alumni["Gender"] column however. You will need to update the column when using the replace command 'alumni["Gender"]=', show how this is done below

```python
alumni['Gender']=alumni['Gender'].replace(to_replace=['M'],value='Male')
alumni
```

```python
alumni
```

## You can set it directly by using the df.loc command, show how this can be done by using the 'df.loc[row_indexer,col_indexer] = value' command to convert the 'M' to 'Male'

```python
alumni.loc[alumni.Gender=='M']='Male'
alumni
```

## Now run the 'value_counts' for Gender again to see the correct columns - 'Male' and 'Female'

```python
alumni['Gender'].value_counts()
```
